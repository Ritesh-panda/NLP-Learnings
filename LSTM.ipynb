```python
input_seq=[]
for sentence in text.split('\n'):
  tkn_sentence=tkn.texts_to_sequences([sentence])

  for i in range(1, len(tkn_sentence[0])):
    input_seq.append(tkn_sentence[0][:i+1])
input_seq
```

```python
#these numbers are not in the same dimensions so due to which we have to add the zero pading
#which will help in getting the dimension same
#zero padding:- 0 in back and front
#get the larget line
max_len=max([len(x) for x in input_seq])
max_len
```

```python
from tensorflow.keras.preprocessing.sequence import pad_sequences
padded=pad_sequences(input_seq, maxlen=max_len, padding='pre')
padded
```

```python
#so in the pad seq the last will be y and the first will be x
x=padded[:,:-1]
y=padded[:,-1]
```

###Model building

```python
from tensorflow.keras.layers import Embedding, LSTM, Dense
```

```python
model=Sequential()
model.add(Embedding(102, 100, input_length=max_len-1))#x.shape will be going inside
model.add(LSTM(150))#100 is a hyperparameter and the dimensions of all the ct ft zt will be 100
model.add(Dense(102, activation='softmax'))
```

```python
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
model.summary()
```

```python
model.build(input_shape=(None, x.shape[1])) # Build the model explicitly
model.summary()
```

Here's a breakdown of the parameters for each layer:

*   **Embedding Layer**: This layer converts word indices into dense vectors. The number of parameters is calculated as `vocabulary_size * embedding_dimension`.
    *   `vocabulary_size` is 102 (from `num_classes` in `Embedding(102, 100, ...)`).
    *   `embedding_dimension` is 100.
    *   Parameters: `102 * 100 = 10,200`

*   **LSTM Layer**: The parameters for an LSTM layer with `units` are calculated as `4 * ((input_dim + units) * units + units)`.
    *   `input_dim` here is the `embedding_dimension` from the previous layer, which is 100.
    *   `units` for the LSTM layer is 100.
    *   Parameters: `4 * ((100 + 100) * 100 + 100) = 4 * (200 * 100 + 100) = 4 * 20,100 = 80,400`

*   **Dense Layer**: This is a standard fully connected layer. The number of parameters is `input_dim * units + units` (for biases).
    *   `input_dim` here is the number of units from the LSTM layer, which is 100.
    *   `units` for the Dense layer is 102.
    *   Parameters: `100 * 102 + 102 = 10,200 + 102 = 10,302`

**Total parameters**: `10,200 (Embedding) + 80,400 (LSTM) + 10,302 (Dense) = 100,902`.

## Using Pre-trained GloVe Embeddings to Improve Accuracy

First, let's download the pre-trained GloVe embeddings. We'll use a relatively small version (50-dimensional vectors, trained on Wikipedia 2014 + Gigaword 5 corpus) to keep the download and processing time manageable for this example.

```python
import os
import zipfile
import requests

# URL for GloVe embeddings (50d, Wikipedia 2014 + Gigaword 5)
glove_url = "http://nlp.stanford.edu/data/glove.6B.zip"
glove_zip_path = "glove.6B.zip"
glove_dir = "glove.6B"

# Download the zip file if it doesn't exist
if not os.path.exists(glove_zip_path):
    print(f"Downloading {glove_url}...")
    response = requests.get(glove_url, stream=True)
    with open(glove_zip_path, "wb") as f:
        for chunk in response.iter_content(chunk_size=8192):
            f.write(chunk)
    print("Download complete.")

# Unzip the file if the directory doesn't exist
if not os.path.exists(glove_dir):
    print(f"Unzipping {glove_zip_path}...")
    with zipfile.ZipFile(glove_zip_path, 'r') as zip_ref:
        zip_ref.extractall(glove_dir)
    print("Unzipping complete.")

print(f"GloVe embeddings are in: {glove_dir}")
```

```python
embeddings_index = {}
f = open(os.path.join(glove_dir, 'glove.6B.50d.txt'))
for line in f:
    values = line.split()
    word = values[0]
    coefs = np.asarray(values[1:], dtype='float32')
    embeddings_index[word] = coefs
f.close()
print(f'Found {len(embeddings_index)} word vectors.')
```

```python
embedding_dim = 50 # GloVe 50-dimensional vectors
vocab_size = len(tkn.word_index) + 1 # +1 for 0-padding in Keras Embedding layer
embedding_matrix = np.zeros((vocab_size, embedding_dim))

for word, i in tkn.word_index.items():
    embedding_vector = embeddings_index.get(word)
    if embedding_vector is not None:
        # Words not found in embedding index will be all-zeros.
        embedding_matrix[i] = embedding_vector
print(f'Shape of embedding matrix: {embedding_matrix.shape}')
```

\Now, we will recreate the model, but this time we'll initialize the `Embedding` layer with the `embedding_matrix` we just created. We'll also set `trainable=False` for the embedding layer, which means these pre-trained weights will not be updated during our training process. This is a common practice when using pre-trained embeddings with small datasets to prevent them from being distorted.

```python
model_glove = Sequential()
model_glove.add(Embedding(vocab_size, # Use vocab_size from embedding_matrix

                          embedding_dim, # Use embedding_dim from GloVe
                          weights=[embedding_matrix], # Initialize with pre-trained weights
                          input_length=max_len-1,
                          trainable=False)) # Set to False for fine-tuning
model_glove.add(LSTM(100))
model_glove.add(Dense(vocab_size, activation='softmax')) # Use vocab_size for output

model_glove.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
model_glove.summary()
```

```python
model_glove.fit(x, to_categorical(y, num_classes=vocab_size), epochs=100)
```
