```python
import nltk
from nltk.tokenize import word_tokenize
from nltk import ngrams
```

```python
nltk.download('punkt_tab')
```

```python
def g_n_grams(text, n):
  words=word_tokenize(text)
  n_grams=list(ngrams(words,n))
  return n_grams
```

```python
txt="i have the many sentences but the ngram will predict the sentence acording to the previos words"
uni=g_n_grams(txt,1)
bi=g_n_grams(txt,2)
tri=g_n_grams(txt,3)
print(uni)
print(bi)
print(tri)
```
