##Special char Removal


```python
import re
```

```python
str="hello how$ are# you*"
clean_Str=re.sub(r"[^a-zA-Z0-9\s]","",str)
print(clean_Str)
```

```python
str1="helloh5876765864 how$ are# you*"
clean_Str1=re.sub(r"[^a-zA-Z\s]","",str1)
print(clean_Str1)
```
