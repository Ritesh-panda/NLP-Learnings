```python
from sklearn.feature_extraction.text import TfidfVectorizer
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
```

```python
df=pd.read_csv("/content/Ecommerce_data.csv", engine='python', on_bad_lines='skip')
```

```python
df.head()
```

```python
df.label.value_counts()
```

```python
corpus=[
    "Thor eating pizza, loki is eating pizza, ironman ate pizza already",
"google is announcing new pixel 6 tomorrow",
"apple is announcing new iphone tomorrow",
"microsoft is announcing new surface tomorrow",
"tesla is announcing new model 3 tomorrow",
"amazon is announcing new echo-dot tomorrow"
"i am eating biriyani and you are eating grapes"]
```

```python
v=TfidfVectorizer()
v.fit(corpus)
transform_output=v.transform(corpus)
```

```python
print(v.vocabulary_)
```

```python
#print idf for each word
all_feature_names=v.get_feature_names_out()
for word in all_feature_names:
  indx=v.vocabulary_.get(word)
  idf_score=v.idf_[indx]
  print(f"{word}:{idf_score}")
```

```python
df['label_num'] = df['label'].map({
    'Household':0,
'Electronics':1	,
'Clothing & Accessories':2	,
'Books':3
})
```

```python
#train test spilt
X_train,X_test,y_train,y_test=train_test_split(df.Text,df.label_num,test_size=0.2 , random_state=42)
```

###Vectorizer

```python
v1=TfidfVectorizer()
x_train_tf=v1.fit_transform(X_train)
x_test_tf=v1.transform(X_test)
```

###classification model

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
model=DecisionTreeClassifier()
model.fit(x_train_tf,y_train)
model.predict(x_test_tf)
model.score(x_test_tf,y_test)
```

```python
print(classification_report(y_test,model.predict(x_test_tf)))
```
