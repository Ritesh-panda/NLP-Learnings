```python
from gensim.test.utils import datapath
import gensim.downloader as api
from scipy.stats import spearmanr
import numpy as np
import pandas as pd
```

```python
#load the pretrained embedded using
glove=api.load("glove-twitter-200")#200 dimensional space vector
fasttext=api.load("fasttext-wiki-news-subwords-300")#300 dimensional space vector
```

```python
print("glove loaded:vocab size", len(glove.key_to_index))
print("fasttext loaded:vocab size" ,len(fasttext.key_to_index))
```

###use case : word similarity task on wordsim-353 dataset

```python
wordsim_path=datapath("/content/wordsim353crowd.csv")
```

```python
print("Evaluation of WordSim353")

# The wordsim_path variable holds the path to the wordsim353crowd.csv file.
# We pass this path directly to evaluate_word_pairs and specify the delimiter as ','
# to ensure it correctly parses the CSV file.

glove_wordsim = glove.evaluate_word_pairs(wordsim_path, delimiter=',')
fasttext_wordsim = fasttext.evaluate_word_pairs(wordsim_path, delimiter=',')

print("glove pearson correlation:", glove_wordsim[0][0])
print("glove spearman correlation:", glove_wordsim[1].correlation)
print("glove oov ratio:", glove_wordsim[2])

print("fasttext pearson correlation:", fasttext_wordsim[0][0])
print("fasttext spearman correlation:", fasttext_wordsim[1].correlation)
print("fasttext oov ratio:", fasttext_wordsim[2])
```
