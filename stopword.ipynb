```python
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize, sent_tokenize
nltk.download('punkt_tab')
nltk.download('stopwords')
```

```python
s="this is a sample token which will be showing off the stop words filteration"
words=word_tokenize(s)
words
```

```python
new=[word for word in words if word.lower() not in stopwords.words('english')]
```
